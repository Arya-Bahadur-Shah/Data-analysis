```python
import matplotlib.pyplot as plt
def StatisticsSummary_(dataframe):
    numeric_columns = dataframe.select_dtypes(include=['float64', 'int64'])

    summary = pd.DataFrame({
        'Sum': numeric_columns.sum(),
        'Mean': numeric_columns.mean(),
        'Standard Deviation': numeric_columns.std(),
        'Skewness': numeric_columns.skew(),
        'Kurtosis': numeric_columns.kurt()
    })
    print("\nSummary Statistics:\n")
    print(summary)

StatisticsSummary_(dataset)
```

```python
def print_unique_values(dataset):
    for column in dataset.columns:
        unique_vals = dataset[column].unique()
        print(f"\nUnique values in column '{column}':")
        print(unique_vals)
print_unique_values(dataset)
```

```python
def View_correlation(dataframe):
    numericColumns = dataframe.select_dtypes(include=['float64', 'int64'])
    correlationValues = numericColumns.corr()

    print("\nCorrelation Matrix:\n")
    print(correlationValues)

View_correlation(dataset)
```

```python

# Identify the 5 least common complaint types
least_5_types = dataset['Complaint Type'].value_counts().nsmallest(5).index

# Filter the DataFrame
filtered_df = dataset[dataset['Complaint Type'].isin(least_5_types)]

# Group by month and complaint type (using 'ME' instead of deprecated 'M')
monthly_trend = filtered_df.groupby([pd.Grouper(key='Created Date', freq='ME'), 'Complaint Type']).size().unstack(fill_value=0)

# Plot
plt.figure(figsize=(12, 6))
for complaint in monthly_trend.columns:
    plt.plot(monthly_trend.index, monthly_trend[complaint], marker='o', label=complaint)

plt.title("Monthly Trend of 5 Least Common Complaint Types")
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.legend(title="Complaint Type")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Group by Complaint Type and Borough, calculate average closing time in hours
grouped = dataset.groupby(['Complaint Type', 'Borough'])['Request_Closing_Hours'].mean().unstack()

# Optionally, filter to top N complaint types with most entries
top_complaints = dataset['Complaint Type'].value_counts().head(6).index
filtered_grouped = grouped.loc[top_complaints]

# Plot grouped bar chart
filtered_grouped.plot(kind='bar', figsize=(12, 6), colormap='tab10', edgecolor='black')

plt.title("Average Request Closing Time by Complaint Type and Borough")
plt.ylabel("Avg. Request Closing Time (Hours)")
plt.xlabel("Complaint Type")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Borough')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
from scipy.stats import f_oneway
required_data = dataset[['Complaint Type', 'Request_Closing_Time']]
unique_complaints = required_data['Complaint Type'].unique()
groups = []
for complaint_types in unique_complaints:
    matched_rows = required_data['Complaint Type'] == complaint_types
    complaint_data = required_data[matched_rows]
    closing_times = complaint_data ['Request_Closing_Time'].dt.total_seconds()
    groups.append(closing_times)

f_stat, p_val = f_oneway(*groups)
print("F-Statistic:", f_stat)
print("P-Value:", p_val)
if p_val < 0.05:
    print("Reject Null Hypothesis: There is a difference in the average response time across complaint types.")
else:
    print("Fail to Reject Null Hypothesis: There is not difference in average response time accross complaint types.")
    
```
